```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from statsmodels.tsa.stattools import adfuller
from scipy.stats import shapiro
from statsmodels.graphics.gofplots import qqplot
```

```python
lst = []
for i in dfun:
    lst.append(i)
lst
```

```python
all_combs = []
for i in range(len(lst)):
    for x in lst[i:]:
        if lst[i] != x:
            all_combs.append((lst[i], x))
all_combs
```

```python
len(all_combs)
```

```python
fig = plt.figure(figsize=(16, 16))
for i in range(len(all_combs)):
    x, y = all_combs[i]
    fig.add_subplot(9, 5, i+1).scatter(dfun[x], dfun[y])
plt.show()
```

```python
fig = plt.figure(figsize=(25, 50))
for i in range(len(all_combs)):
    x, y = all_combs[i]
    fig.add_subplot(15, 3, i+1).hexbin(dfun[x], dfun[y])
plt.show()
```
